```python
import pandas as pd
from collections import defaultdict
```

```python
lines = None
with open('functions_cozmo.txt', 'r') as file:
    lines = file.readlines()
```

```python
with open("functions_cozmo.txt",'w') as file:
    for i,line in enumerate(lines,1):         ## STARTS THE NUMBERING FROM 1 (by default it begins with 0)    
        if line.strip() == '-':                              ## OVERWRITES line:2
            file.writelines("")
        else:
            file.writelines(line)
```

```python
linesByID = defaultdict(list)
```

```python
for line in lines:
    id = line.split(' ')[0]
    linesByID[id].append(' '.join(line.split(' ')[1:]))
    print('id: ', id)
    print('function: ', ' '.join(line.split(' ')[1:]))
```

```python
print(len(linesByID['1616770284.3265388']))
```
