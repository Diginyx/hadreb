```python
import pandas as pd
import numpy as np
from collections import defaultdict
import pickle
```

```python
def appendParen(x):
    if x != 'nan':
        return '(' + str(x)
    else:
        return x
```

```python
misty_functions = pd.read_csv(r'functions_misty.txt', sep='(')
misty_functions.to_csv (r'functions_misty.csv', index = None)
misty_functions
```

```python
misty_functions['Function'] = misty_functions['Function'].apply(lambda x: appendParen(x))
```
